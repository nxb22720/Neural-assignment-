```python
# Part 1: String Manipulation
# Input the string from console
input_str = input("Enter the string 'Python': ")
char_list = list(input_str)

# Delete at least 2 characters (e.g., the last two)
if len(char_list) >= 2:
    char_list.pop()
    char_list.pop()
# Reverse the resultant string
char_list.reverse()
result_str = "".join(char_list)
print("Resultant string:", result_str)
```

```python
# Part 2: Arithmetic Operations
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print(f"Addition: {num1 + num2}")
print(f"Subtraction: {num1 - num2}")
print(f"Multiplication: {num1 * num2}")
print(f"Division: {num1 / num2 if num2 != 0 else 'Undefined'}")
```
